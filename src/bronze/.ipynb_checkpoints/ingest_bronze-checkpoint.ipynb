# Read local JSON file

##### <font color="green"> Import library and define file source </font>

```python
import pandas as pd

file = '../../data/bronze/bronze_issues.json'
```

##### <font color="green"> Read the JSON file. It was necessary to set the read type to Series, because setting it to Frame was returning an error </font>

```python
read_series = pd.read_json(file, typ='series')
read_series
```

##### <font color="green"> Get list of key issues </font>

```python
df = pd.json_normalize(read_series.loc['issues'])
df.head(2)
```

##### <font color="green"> Save bronze table </font>

```python
df.to_parquet("../../data/bronze/bronze_issues.parquet", index=False)
```

Por quê a escolha do Parquet?

Formato colunar (melhor performance para leitura analítica)

Compactação eficiente (menor tamanho de arquivo)

Mantém schema e tipos complexos (listas/dicts como no seu assignee e timestamps)

Padrão de mercado em Data Lake (Databricks, AWS, Azure, GCP)

Ideal para pipelines de dados modernos
